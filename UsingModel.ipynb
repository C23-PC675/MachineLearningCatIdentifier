```python
from tensorflow.keras.models import model_from_json

# Load JSON
with open('/kaggle/input/modelcatbreed/model.json', 'r') as json_file:
    loaded_model_json = json_file.read()
model = model_from_json(loaded_model_json)

# Load weights
model.load_weights('/kaggle/input/modelcatbreed/model.h5') 
```

```python
from tensorflow import keras

# Load the H5 model
model = keras.models.load_model('/kaggle/input/modelcatbreed/model.h5')
```

```python
model.summary()
```

```python
import cv2
import numpy as np
from tensorflow.keras.preprocessing import image
import time

image_size = (224, 224)

# Load Image
image_path = '/kaggle/input/bengalcat/karakteristik-kucing-bengal-dan-cara-merawatnya-perlu-diketahui.jpg'  # Image Path
img = cv2.imread(image_path)
img = cv2.resize(img, image_size)
img = img.astype('float32') / 255.0
img = np.expand_dims(img, axis=0)

# Predict using model and save time to predict
start_time = time.time()
predictions = model.predict(img)
end_time = time.time()
prediction_time = end_time - start_time

# Define label manually
class_labels = ["Bengal", "Bombay", "Persian", "Torbie", "Tuxedo"]

# Print the predicted class labels, confidences as percentages, and prediction time
for i, prediction in enumerate(predictions[0]):
    class_label = class_labels[i]
    confidence = prediction * 100
    print("Class:", class_label)
    print("Confidence:", confidence, "%")

print("Prediction Time:", prediction_time, "seconds")
```

```python
import cv2
import numpy as np
from tensorflow.keras.preprocessing import image
import time

image_size = (224, 224)

# Load Image
image_path = '/kaggle/input/bengalcat/karakteristik-kucing-bengal-dan-cara-merawatnya-perlu-diketahui.jpg'  # Image Path
img = cv2.imread(image_path)
img = cv2.resize(img, image_size)
img = img.astype('float32') / 255.0
img = np.expand_dims(img, axis=0)

# Make predictions
start_time = time.time()
predictions = model.predict(img)
end_time = time.time()
prediction_time = end_time - start_time

# Class labels
class_labels = ["Bengal", "Bombay", "Persian", "Torbie", "Tuxedo"]

# Print predicted class, confidences, and prediction time
for i, prediction in enumerate(predictions[0]):
    class_label = class_labels[i]
    confidence = prediction * 100
    print("Class:", class_label)
    print("Confidence:", confidence, "%")

# Get highest confidence
predicted_class_index = np.argmax(predictions[0])
predicted_class_label = class_labels[predicted_class_index]
highest_confidence = predictions[0][predicted_class_index] * 100

# Print highest confidence
print("")
print("The Cat Is", predicted_class_label, "With Highest Confidence:", highest_confidence, "%")
print("Prediction Time:", prediction_time, "seconds")
```
